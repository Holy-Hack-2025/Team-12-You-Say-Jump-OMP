```python
import yfinance as yf
import pandas as pd

# List of five American companies (by ticker symbols)
tickers = ['AAPL', 'MSFT', 'TSLA', 'AMZN', 'GOOGL']

# Initialize an empty dictionary to store ESG data
esg_data = {}

# Loop through the list of tickers and fetch ESG data
for ticker in tickers:
    company = yf.Ticker(ticker)
    try:
        # Fetch sustainability data
        sustainability_data = company.sustainability
        print(f"ESG data for {ticker}:")
        print(sustainability_data.totalEsg)  # Print the raw data
        if sustainability_data.empty:
            esg_data[ticker] = {'Environmental': None, 'Social': None, 'Governance': None}
        else:
            esg_data[ticker] = sustainability_data.iloc[0].to_dict()  # Get first row (current data)
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        esg_data[ticker] = {'Environmental': None, 'Social': None, 'Governance': None}

# Convert the dictionary to a DataFrame
esg_df = pd.DataFrame(esg_data).transpose()

# Save the DataFrame to an Excel file
esg_df.to_excel('American_Companies_ESG_Scores.xlsx', engine='openpyxl')

# Print out the DataFrame to verify
print(esg_df)
```

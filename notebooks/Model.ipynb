```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    """Function that starts high, decreases to a minimum, and slightly increases at high x."""
    a, b, c = 1, -4, 5  # Coefficients for the quadratic part y = ax^2 + bx + c
    y = a*x**2 + b*x + c  # Quadratic decreasing function

    # At high x, we slightly increase the function
    threshold = 4  # Cutoff point where we modify behavior
    if x > threshold:
        y += 0.1 * (x - threshold)  # Small linear increase

    return y

# Generate values for visualization
x_values = np.linspace(0, 10, 100)
y_values = np.array([f(x) for x in x_values])

# Plot the function
plt.plot(x_values, y_values, label="Custom Function")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Function that decreases and then slightly increases")
plt.axvline(x=4, linestyle="--", color="red", label="Cutoff Point")
plt.legend()
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the function with reflection across x = 4
def f(x):
    """Function that starts high, decreases to a minimum, and slightly increases at high x."""
    a, b, c = 1, -4, 5  # Quadratic coefficients
    x_mirrored = 8 - x  # Reflect x across x = 4
    y = a * x_mirrored**2 + b * x_mirrored + c  # Apply quadratic formula

    # Smoothly increasing noise function for x > 6
    base_noise = 0.5
    extra_noise = 2 * ((x - 6) / 4) if x > 6 else 0  # Gradual increase from x = 6 to 10
    noise = np.random.normal(0, base_noise + extra_noise)

    return y + noise

# Generate grid data to create a surface
x_values = np.linspace(0, 10, 100)
y_values = np.array([f(x) for x in x_values])

# Ensure y and z are highly correlated by adding very small noise to y-values
z_values = y_values + np.random.normal(0, 0.3, len(y_values))  # Small noise keeps correlation high

# Create a meshgrid for surface representation
X, Y = np.meshgrid(x_values, x_values)
Z = np.array([[f(x) for x in row] for row in X])  # Compute Z values for the entire grid

# Create an interactive 3D surface plot
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='Viridis')])

# Add axis labels and title
fig.update_layout(
    title="Interactive 3D Surface Plot",
    scene=dict(
        xaxis_title="X-axis",
        yaxis_title="Y-axis",
        zaxis_title="Z-axis"
    )
)

# Show interactive plot
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the function for a convex shape with a clear minimum
def f_cone(x, y):
    """Creates a convex, coned shape with a clear minimum at (x=5, y=5)."""
    # Paraboloid shape: z increases as (x, y) move away from the center
    z = (x - 5)**2 + (y - 5)**2  # Minimum at (5,5)

    # Add mild noise to simulate real-world smooth convexity
    noise = np.random.normal(0, 0.2)  # Small noise to keep smoothness
    return z + noise

# Generate a grid for the convex surface
x_values = np.linspace(0, 10, 100)
y_values = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x_values, y_values)
Z = np.array([[f_cone(x, y) for x, y in zip(row_x, row_y)] for row_x, row_y in zip(X, Y)])

# Create an interactive 3D surface plot
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='Viridis')])

# Add axis labels and title
fig.update_layout(
    title="Interactive 3D Convex Surface (Coned Shape)",
    scene=dict(
        xaxis_title="X-axis",
        yaxis_title="Y-axis",
        zaxis_title="Z-axis"
    )
)

# Show interactive plot
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Define the function to create an asymmetric convex shape with a sharp edge at the bottom
def f_convex(x, y):
    """Creates an asymmetric convex shape with a clear minimum and sharp bottom edge."""
    # Asymmetric quadratic function: different curvature in x and y
    z = (x - 3)**2 + 1.5 * (y - 6)**2  # Minimum at (x=3, y=6)

    # Add a sharper edge at the bottom to prevent flattening
    z += np.exp(-0.8 * ((x - 3)**2 + (y - 6)**2)) * 4  # Creates a sharper edge at the min

    # Add mild noise to simulate real-world data smoothness
    noise = np.random.normal(0, 0.2)  # Small noise for smooth convexity
    return z + noise

# Generate grid data to create a surface
x_values = np.linspace(0, 10, 100)
y_values = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x_values, y_values)
Z = np.array([[f_convex(x, y) for x, y in zip(row_x, row_y)] for row_x, row_y in zip(X, Y)])

# Create an interactive 3D surface plot
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='Viridis')])

# Add axis labels and title
fig.update_layout(
    title="Interactive 3D Asymmetric Convex Surface with Sharp Bottom",
    scene=dict(
        xaxis_title="X-axis",
        yaxis_title="Y-axis",
        zaxis_title="Z-axis"
    )
)

# Show interactive plot
fig.show()
```

```python
import numpy as np
import casadi as ca
import plotly.graph_objects as go

# Define the function in CasADi
x = ca.SX.sym("x")
y = ca.SX.sym("y")

# Rotate the minimum to a higher x value (shifted to x=7 instead of x=3)
x_rotated = 10 - x  # Rotation along a vertical axis

# Asymmetric quadratic function: different curvature in x and y
z = (x_rotated - 7)**2 + 1.5 * (y - 6)**2  # Minimum at (x=7, y=6)

# Add a sharper edge at the bottom
z += ca.exp(-0.8 * ((x_rotated - 7)**2 + (y - 6)**2)) * 4  

# Define the optimization problem
f = ca.Function("f", [x, y], [z])  # Define function

# Define optimization variables
opt_variables = ca.vertcat(x, y)

# Compute gradient (first derivative)
grad = ca.gradient(z, opt_variables)

# Compute Hessian (second derivative matrix)
hessian = ca.hessian(z, opt_variables)[0]

# Define solver using Newton's method
solver = ca.nlpsol("solver", "ipopt", {"x": opt_variables, "f": z, "g": grad})

# Initial guess
initial_guess = [5, 5]  # Starting from the center of the range

# Solve optimization problem
sol = solver(x0=initial_guess)

# Extract optimal values
optimal_x, optimal_y = sol["x"].full().flatten()
optimal_z = float(f(optimal_x, optimal_y).full())  # Convert to float

print(f"Optimal Solution: x = {optimal_x:.4f}, y = {optimal_y:.4f}, z = {optimal_z:.4f}")

# Generate grid data for visualization
x_values = np.linspace(0, 10, 100)
y_values = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x_values, y_values)
Z = np.array([[f(x, y).full().flatten()[0] for x, y in zip(row_x, row_y)] for row_x, row_y in zip(X, Y)])

# Create an interactive 3D surface plot
fig = go.Figure(data=[go.Surface(x=X, y=Y, z=Z, colorscale='Viridis')])

# Plot the found minimum point
fig.add_trace(go.Scatter3d(
    x=[optimal_x], y=[optimal_y], z=[optimal_z],
    mode='markers', marker=dict(size=8, color='red'),
    name='Optimal Point'
))

# Add axis labels and title
fig.update_layout(
    title="Optimized Minimum of Rotated Asymmetric Convex Surface",
    scene=dict(
        xaxis_title="X-axis",
        yaxis_title="Y-axis",
        zaxis_title="Z-axis"
    )
)

# Show interactive plot
fig.show()
```
